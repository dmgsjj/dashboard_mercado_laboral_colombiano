Objetivo: inspeccionar estructura real de las bases, verificar variables clave, revisar valores faltantes y distribuciones.

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
from src.config import VARS_DASHBOARD, DTYPE_MAP, FACTOR_EXPANSION
from src.loaders import cargar_año
```

```python
# Cargar un año para inspección (solo columnas del dashboard)
AÑO = 2022
df = cargar_año(AÑO, columnas=None)  # None = cargar todo
print(f'Shape: {df.shape}')
df.head(3)
```

```python
# Verificar que las variables clave existen
vars_clave = ['AREA', 'DPTO', 'P3271', 'P6040', 'OCI', 'DSI',
              'RAMA2D_R4', 'P6500', 'FEX_C18', 'MES']
faltantes = [v for v in vars_clave if v not in df.columns]
print('Variables clave faltantes:', faltantes if faltantes else 'NINGUNA ✓')
```

```python
# Tipos de datos
df[vars_clave].dtypes
```

```python
# Valores faltantes en variables clave
df[vars_clave].isna().sum().to_frame('nulos')
```

```python
# Distribución de meses
print('Meses disponibles:', sorted(df['MES'].dropna().unique().tolist()))
df['MES'].value_counts().sort_index()
```

```python
# Valores únicos AREA (ciudades)
print('Códigos AREA:', sorted(df['AREA'].dropna().unique().tolist()))
```

```python
# Distribución OCI y DSI
print('OCI:', df['OCI'].value_counts().to_dict())
print('DSI:', df['DSI'].value_counts().to_dict())
```

```python
# Estadísticas básicas FEX_C18 y P6500
df[[FACTOR_EXPANSION, 'P6500']].describe()
```

```python
# Cálculo rápido TD nacional mes 1 (verificación manual)
mes1 = df[df['MES'] == 1].copy()
pet = mes1[mes1['P6040'] >= 15]
ocu = (pet['OCI'].fillna(0) * pet['FEX_C18']).sum()
des = (pet['DSI'].fillna(0) * pet['FEX_C18']).sum()
pea = ocu + des
print(f'TD enero {AÑO}: {des/pea*100:.1f}%  (referencia DANE: 13.7%)')
```
